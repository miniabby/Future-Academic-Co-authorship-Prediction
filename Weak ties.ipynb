```python
import pandas as pd
import numpy as np
import networkx as nx
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
```

```python
# drop all edges which CN = 0
cond = train[train['CN'] == 0].index
nonCN = train.drop(index=cond, inplace=False)
nonCN
```

```python
# create a graph for nonCN
nodeL = np.union1d(nonCN['Source'].values, nonCN['Sink'].values)
len(nodeL)
```

```python
# create graph
G = nx.from_pandas_edgelist(nonCN, "Source", "Sink", create_using=nx.Graph())
```

```python
cnL = []
for u,v in tqdm(nonCN.Pair):
    cn = sorted(nx.common_neighbors(G,u,v))
    cnL.append(((u,v), list(cn)))
cnDF = pd.DataFrame(cnL, columns=["Pair", "Source", "Sink", "NCA", "cn_list"])
```
